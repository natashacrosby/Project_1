```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from us import states


df = pd.read_csv("poverty.csv")

df.head()
```

```python
df.set_index("GEO.display-label", inplace=True)
df.head()
```

```python
df.drop(df.columns[0:2], inplace = True, axis = 1)
df.drop(df.columns[4:22], inplace = True, axis = 1)
df.head()
```

```python
df.drop(df.columns[1:4], inplace = True, axis = 1)
df.head()
```

```python
renamed_df = df.rename(columns={"HD01_VD01":"Total"})
renamed_df.head()
```

```python
renamed_df.drop(['Geography'], axis = 0, inplace = True)
```

```python
renamed_df.reset_index(drop=False, inplace = True)
```

```python
poverty1_df = renamed_df.rename(columns={"GEO.display-label":"County"})
poverty1_df.head()
```

```python
poverty2 =poverty1_df.replace(regex=[' County'], value='')
poverty2.head()
```

```python
POV3 = poverty2['County'].str.split(pat = ",",expand= True)
poverty2['county']=POV3[0]
poverty2['state']=POV3[1]
poverty2.drop(columns =["County"], inplace = True)
poverty2.head()
```

```python
poverty2['county'] = poverty2['county'].map(str) + ", " + poverty2['state']
poverty2 = poverty2[["Total", "county"]]
poverty2.head()
```

```python
crime_df = pd.read_csv("crime_data.csv")

crime_df.head()
```

```python
crime_df = crime_df[["county_name","population"]]
crime_df= crime_df.rename(columns = {"county_name":"county"})


crime_df['county'].replace(regex=True,inplace=True,to_replace=r' County',value=r'')
crime_df['county'].replace(regex=True,inplace=True,to_replace=r' Parish',value=r'')
crime_df['county'].replace(regex=True,inplace=True,to_replace=r' city',value=r'')
crime_df['county'].replace(regex=True,inplace=True,to_replace=r' Borough',value=r'')
crime_df['county'].replace(regex=True,inplace=True,to_replace=r' Census Area',value=r'')
crime_df.head()
```

```python
poverty_3 = poverty2.merge(crime_df, on="county")
poverty_3.head()
```
