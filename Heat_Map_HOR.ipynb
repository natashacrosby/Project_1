```python
import gmaps
import numpy as np
import pandas as pd
import requests
import time
from us import states
from config1 import gkey
import json
from pprint import pprint
```

```python
home_ownership_csv = pd.read_csv("Homeownership_ACS.clean")
home_ownership_df = pd.DataFrame(home_ownership_csv)
```

```python
home_ownership_df = home_ownership_df[["County", "HOR"]]
home_ownership_df  = home_ownership_df.rename(columns={"County": "county"})
```

```python
home_ownership_df.head()
```

```python
new= home_ownership_df["county"].str.split(",", n = 1, expand = True)
home_ownership_df["county"]= new[0]
home_ownership_df["state"]= new[1]
home_ownership_df.head()
```

```python
home_ownership_df["Lat"]=""
home_ownership_df["Long"]=""
home_ownership_df.head()
```

```python
params = {"key": gkey}
for index, row in home_ownership_df.iterrows():
    try:
        base_url = "https://maps.googleapis.com/maps/api/geocode/json"

        county = row["county"]
        state = row["state"]
        # update address key value
        params['address'] = f"{county},{state}"

        # make request
        county_lat_lng = requests.get(base_url, params=params)
    
        county_lat_lng = county_lat_lng.json()

        home_ownership_df.loc[index, "Lat"] = county_lat_lng["results"][0]["geometry"]["location"]["lat"]
        home_ownership_df.loc[index, "Lng"] = county_lat_lng["results"][0]["geometry"]["location"]["lng"]
        print(f'processing city {index}')
        time.sleep(1)
    except IndexError:
        print(f"City not found. Skipping...")
        home_ownership_df.loc[index, 'Lat']= None
home_ownership_df.head()
```

```python
home_ownership_df.to_csv("Home_ownership_Geo_Info.csv")
```

## Heatmap of HOR

```python
home_ownership_df = home_ownership_df.dropna()
```

```python
gmaps.configure(api_key=gkey)
```

```python
locations = home_ownership_df[["Lat", "Lng"]].astype(float)
HOR = home_ownership_df["HOR"].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=HOR, 
                                 dissipating=False, max_intensity=200,
                                 point_radius = 1)

# Adjust heat_layer setting to help with heatmap dissipating on zoom
heat_layer.dissipating = False
heat_layer.max_intensity = 200
heat_layer.point_radius = .5

fig.add_layer(heat_layer)

fig
```
