```python
state_df = pd.DataFrame(list(us_state_abbrev.items()), columns=['State', 'State_abbrev'])
state_df.head()
```

```python
df.reset_index(inplace=True, drop=False)
df.head()
```

```python
test = pd.merge(df, state_df, how="outer", left_on="Counties", right_on = "State", sort = False)
```

```python
test.head()
```

```python
test = test.sort_values(by="index")
test.head()
```

```python
test = test.fillna(method='ffill')
test.head()
```

```python
test = test.reset_index(drop = True)
test.head()
```

```python
for index, row in test.iterrows():
    if test.loc[index, 'Counties']== test.loc[index, 'State_y']:
        test.loc[index, 'State_y'] = np.nan
test.head()
```

```python
income_data_clean = test[["Counties", "2015","State_y","State_abbrev"]]
income_data_clean.head()
```

```python
income_data_clean = income_data_clean.dropna(how="any")
income_data_clean.head()
```

```python
income_data_clean = income_data_clean[["Counties", "2015","State_abbrev"]]
income_data_clean.head()
```

```python
income_data_clean = income_data_clean.reset_index(drop = True)
```

```python
income_data_clean = income_data_clean.rename(columns={"Counties":"county"})
income_data_clean.head()
```

```python
income_data_clean["county"] = income_data_clean["county"].map(str) + ", "+ income_data_clean["State_abbrev"]
income_data_clean.head()
```

```python
income_data_clean = income_data_clean[["county", "2015"]]
income_data_clean.head()
```
